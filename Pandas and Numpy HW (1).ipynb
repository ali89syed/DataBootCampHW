```python
### Age Demographics


bins = [0, 9.90, 14.90, 19.90, 24.90, 29.90, 34.90, 39.90, 44.90]
group_name = ['0 - 10', '10 - 14', '15 - 19', '20 - 24', '25 - 29', '30 - 34', '35 - 39', '40 - 45']

df['Age_Group'] = pd.cut(df['Age'],bins, labels=group_name)

df.head()

### The below each broken into bins of 4 years (i.e. &lt;10, 10-14, 15-19, etc.)
 ## Purchase Count
zero_ten = df["Age_Group"].value_counts()['0 - 10']
ten_fourteen= df["Age_Group"].value_counts()['10 - 14']
fifteen_nineteen= df["Age_Group"].value_counts()['15 - 19']
twenty_twenty4= df["Age_Group"].value_counts()['20 - 24']
twenty5_twenty9= df["Age_Group"].value_counts()['25 - 29']
thirty_thirty4= df["Age_Group"].value_counts()['30 - 34']
thirty5_thirty9 = df["Age_Group"].value_counts()['35 - 39'] 
forty_forty5 = df["Age_Group"].value_counts()['40 - 45']


df.head()



 ## Average Purchase Price 
  ## Total Purchase Value
 ## Average Purchase Total per Person by Age Group   
    
APPAGEG = df.groupby('Age_Group').mean()['Price']
APPAGEG.head()

TPVAGEG = df.groupby('Age_Group').sum()['Price']
TPVAGEG

tAGEG= df["Age_Group"].value_counts()
tAGEG    

AGEGDF = ({'PurCount': tAGEG, "AvgPurPrice" : APPAGEG, "TotalPurValue" :tAGEG })
AGEGDF

FinalSNDF = pd.DataFrame(AGEGDF)
FinalSNDF.head()
```
